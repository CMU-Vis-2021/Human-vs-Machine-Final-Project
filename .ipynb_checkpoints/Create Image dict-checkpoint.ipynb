```python
import os

img_dict = {}

directory = r'./assets/style-transfer-preprocessed-512/'
for filename in os.listdir(directory):
    print(filename)
    
    if filename not in img_dict:
        img_dict[filename] = {}
    
    newdir = os.path.join(directory, filename)
    for file in os.listdir(newdir):
        
        print(file)
        
        firstnumber = re.search(r'\d+', file).group()
        splitfile = file.split('-')
        splitfile = splitfile[1]
        print(splitfile)
        
        if firstnumber not in img_dict[filename]:
            img_dict[filename][firstnumber] = []
            img_dict[filename][firstnumber].append(splitfile)
        else:
            img_dict[filename][firstnumber].append(splitfile)
```

```python
import json

json_object = json.dumps(img_dict, indent = 4) 
print(json_object)
```
